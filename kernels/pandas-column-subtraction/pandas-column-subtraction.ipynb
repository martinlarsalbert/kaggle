This notebook gives an example how you can substract a column from all columns in a DataFrame in a very elegant way.

## Example data frame

```python
data = np.random.random(size=(5,3))
df = pd.DataFrame(data = data, columns=['A','B','C'])
df
```

```python
df.plot(style='o-')
```

## Data aggregation

Along the columns:

```python
df.mean(axis=0)
```

Along the rows:

```python
df.mean(axis=1)
```

## Math between the column and rows
The data frame aggregation methods as used above is quite standard procedure. What I have not thought about before is that you can use a similar approach for other operations such as addition (**.add**) and subtraction (**.sub**) etc.

Imagine that we want to substract column A from all columns in the data frame, how would you do that? Before I used to do something ugly as for instance:

```python
df_diff_A = pd.DataFrame()
for key in df:
    df_diff_A[key] = df[key] - df['A']
df_diff_A
```

As a more cleaner solution you could image doing something like:

```python
df - df['A']
```

...but this clearly does not workout as hoped. We are here using the **.sub** method:

```python
df.sub(df['A'])
```

And this method has an argument **axis** similarly to **.mean** above. So if we set **axis=0** the operation is now working as how we want it:

```python
df_diff_A = df.sub(df['A'], axis=0)
df_diff_A
```

```python
df_diff_A.plot(style='o-')
```
